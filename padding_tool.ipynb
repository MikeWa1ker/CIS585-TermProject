```python
import pathlib
import tensorflow as tf
```

```python
#This will take the images in the folder_URL and apply a uniform padding that will accomodate
#both landscape and portrait photos and move them to a "paddedFile" folder. I then take those
#images and put them in the corisponding folder (robin, landon, mike)

#individual image sets
folder_url = r"portrait_family_photos\landon"
#folder_url = r"family_photos\landon"
#folder_url = r"family_photos\robin"

#Images used for testiing
#folder_url = "test_family_photos"

data_dir = pathlib.Path(folder_url)

image_count = len(list(data_dir.glob('*')))
print(image_count)
```

```python
#This will loop through each of the photos found above and apply padding to each, making them all the same size. 

for file_path in data_dir.iterdir():
    image_url = folder_url + "\\" + file_path.name
    
    image = tf.io.read_file(image_url)
    image = tf.image.decode_jpeg(image)

    padded_image = tf.image.pad_to_bounding_box(
        image, offset_height=0, offset_width=0, target_height=2049, target_width=1539
    )

    encoded_image = tf.io.encode_jpeg(padded_image)

    new_image_name = "pad-" + file_path.name
    new_image_name = "C:\\Users\MikeW\\Documents\\PaddedFile\\" + new_image_name

    print(new_image_name)
    
    with open(new_image_name, 'wb') as f:
        f.write(encoded_image.numpy())
    
```
